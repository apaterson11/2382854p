```python
# Standard imports
# Make sure you run this cell!
from __future__ import print_function, division
import numpy as np  # NumPy

%matplotlib notebook
%matplotlib notebook

# Set up Matplotlib
import matplotlib as mpl   
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
print("Everything imported OK")
```

```python
## LOSS - DATAGRAMS
# air = np.loadtxt('cherry_trees.txt')
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
# print(data[:,1])
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 6])
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,2], label="Actual Latency")
ax.plot(data[:,0], data[:,3], label="Perceived Latency")
ax.set_title("Datagrams - Adjusted Link Loss vs Latency", fontsize=13)
ax.set_xlabel("Link Loss (percentage)")
ax.set_ylabel("Latency (seconds)")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])

#
#########################

# small_data = np.loadtxt(open("datagrams/small_results.csv", "rb"), delimiter=",", skiprows=1)
# fig = plt.figure()
# axes = plt.gca()
# axes.set_ylim([0, 130])
# ax2 = fig.add_subplot(1,1,1)
# ax2.plot(small_data[:,0], small_data[:,2])
# ax2.plot(small_data[:,0], small_data[:,3])
# ax2.set_title("Datagrams - Adjusted Bandwidth (Small Measurements) vs Latency")
# ax2.set_xlabel("Bandwidth (Mbits)")
# ax2.set_ylabel("Latency (seconds)")

all_data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
plt.autoscale(True)
ax3 = fig.add_subplot(1,1,1)
ax3.plot(all_data[:,0], all_data[:,5])
ax3.set_title("Datagrams - Adjusted Link Loss vs Application Loss", fontsize=13)
ax3.set_xlabel("Link Loss (percentage)")
ax3.set_ylabel("Application Loss (percentage)")
```

```python
## LOSS - STREAMS
# air = np.loadtxt('cherry_trees.txt')
data = np.loadtxt(open("streams/results.csv", "rb"), delimiter=",", skiprows=1)
# print(data[:,1])
plt.ylim(0,6)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 10])
ax4 = fig.add_subplot(1,1,1)
ax4.plot(data[:,0], data[:,1], label="Actual Latency")
ax4.plot(data[:,0], data[:,2], label="Perceived Latency")
ax4.set_title("Streams - Adjusted Link Loss vs Latency", fontsize=13)
ax4.set_xlabel("Loss (Percentage)")
ax4.set_ylabel("Latency (seconds)")

handles, labels = ax4.get_legend_handles_labels()
ax4.legend(handles[::-1], labels[::-1])

#
#########################

# small_data = np.loadtxt(open("streams/small_results.csv", "rb"), delimiter=",", skiprows=1)
# fig = plt.figure()
# axes = plt.gca()
# axes.set_ylim([0, 130])
# ax = fig.add_subplot(1,1,1)
# ax.plot(small_data[:,0], small_data[:,1])
# ax.plot(small_data[:,0], small_data[:,2])
# ax.set_title("Streams - Adjusted Bandwidth (Small Measurements) vs Latency")
# ax.set_xlabel("Bandwidth (Mbits)")
# ax.set_ylabel("Latency (seconds)")
```

```python
## LOSS - DATAGRAMS
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
data_streams = np.loadtxt(open("streams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 10])
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,2], label="Actual Latency (Datagrams)")
ax.plot(data[:,0], data[:,3], label="Perceived Latency (Datagrams)")
ax.plot(data_streams[:,0], data_streams[:,1], label="Actual Latency (Streams)", linestyle='dashed')
ax.plot(data_streams[:,0], data_streams[:,2], label="Perceived Latency (Streams)", linestyle='dashed')
ax.set_title("Datagrams & Streams - Adjusted Link Loss vs Application Latency", fontsize=13)
ax.set_xlabel("Link Loss (percentage)")
ax.set_ylabel("Application Latency (seconds)")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])
```
