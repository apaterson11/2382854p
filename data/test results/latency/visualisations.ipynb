```python
# Standard imports
# Make sure you run this cell!
from __future__ import print_function, division
import numpy as np  # NumPy

%matplotlib notebook
%matplotlib notebook

# Set up Matplotlib
import matplotlib as mpl   
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
print("Everything imported OK")
```

```python
## LATENCY - DATAGRAMS
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 17])
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,2], label="Actual Latency")
ax.plot(data[:,0], data[:,3], label="Perceived Latency")
ax.set_title("Datagrams - Adjusted Link Latency vs Application Latency", fontsize=13)
ax.set_xlabel("Link Latency (milliseconds)")
ax.set_ylabel("Application Latency (seconds)")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])

## LOSS - DATAGRAMS
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,5])
ax.set_title("Datagrams - Adjusted Link Latency vs Packet Loss", fontsize=13)
ax.set_xlabel("Link Latency (milliseconds)")
ax.set_ylabel("Packet Loss (percentage)")
```

```python
## BANDWIDTH - STREAMS
data = np.loadtxt(open("streams/results.csv", "rb"), delimiter=",", skiprows=1)
# plt.ylim(0,6)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 17])
ax4 = fig.add_subplot(1,1,1)
ax4.plot(data[:,0], data[:,1], label="Actual Latency")
ax4.plot(data[:,0], data[:,2], label="Perceived Latency")
ax4.set_title("Streams - Adjusted Link Latency vs Application Latency", fontsize=13)
ax4.set_xlabel("Link Latency (milliseconds)")
ax4.set_ylabel("Application Latency (seconds)")

handles, labels = ax4.get_legend_handles_labels()
ax4.legend(handles[::-1], labels[::-1])
```

```python
## LATENCY - DATAGRAMS
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
data_streams = np.loadtxt(open("streams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 17])
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,2], label="Actual Latency (Datagrams)")
ax.plot(data[:,0], data[:,3], label="Perceived Latency (Datagrams)")
ax.plot(data_streams[:,0], data_streams[:,1], label="Actual Latency (Streams)", linestyle='dashed')
ax.plot(data_streams[:,0], data_streams[:,2], label="Perceived Latency (Streams)", linestyle='dashed')
ax.set_title("Datagrams & Streams - Adjusted Link Latency vs Application Latency", fontsize=13)
ax.set_xlabel("Link Latency (milliseconds)")
ax.set_ylabel("Application Latency (seconds)")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])
ax.legend(fontsize=12.5)
```
