```python
## BANDWIDTH - STREAMS
# air = np.loadtxt('cherry_trees.txt')
data = np.loadtxt(open("streams/results.csv", "rb"), delimiter=",", skiprows=1)
# print(data[:,1])
fig = plt.figure()
axes = plt.gca()
# axes.set_ylim([0, 6])
ax4 = fig.add_subplot(1,1,1)
ax4.plot(data[:,0], data[:,1], label="Actual Latency")
ax4.plot(data[:,0], data[:,2], label="Perceived Latency")
ax4.set_title("Streams - Adjusted Bandwidth vs Latency")
ax4.set_xlabel("Bandwidth (Mbits)")
ax4.set_ylabel("Latency (seconds)")
handles, labels = ax.get_legend_handles_labels()
ax4.legend(handles[::-1], labels[::-1])
#
#########################

small_data = np.loadtxt(open("streams/small_results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 130])
ax = fig.add_subplot(1,1,1)
ax.plot(small_data[:,0], small_data[:,1], label="Actual Latency")
ax.plot(small_data[:,0], small_data[:,2], label="Perceived Latency")
ax.set_title("Streams - Adjusted Bandwidth (Small Measurements) vs Latency")
ax.set_xlabel("Bandwidth (Mbits)")
ax.set_ylabel("Latency (seconds)")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])
```
