```python
# Standard imports
# Make sure you run this cell!
from __future__ import print_function, division
import numpy as np  # NumPy

%matplotlib notebook
%matplotlib notebook

# Set up Matplotlib
import matplotlib as mpl   
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
print("Everything imported OK")
```

```python
## BANDWIDTH - DATAGRAMS
# air = np.loadtxt('cherry_trees.txt')
data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
# print(data[:,1])
fig = plt.figure()
axes = plt.gca()
# axes.set_ylim([0, 6])
ax = fig.add_subplot(1,1,1)
ax.plot(data[:,0], data[:,2], label="Actual Latency")
ax.plot(data[:,0], data[:,3], label="Perceived Latency")
ax.set_title("Datagrams - Adjusted Bandwidth vs Latency", fontsize=13)
ax.set_xlabel("Bandwidth (Mbps)")
ax.set_ylabel("Latency (seconds)")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1])
#
#########################

small_data = np.loadtxt(open("datagrams/small_results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
axes = plt.gca()
axes.set_ylim([0, 130])
ax2 = fig.add_subplot(1,1,1)
ax2.plot(small_data[:,0], small_data[:,2], label="Actual Latency")
ax2.plot(small_data[:,0], small_data[:,3], label="Perceived Latency")
ax2.set_title("Datagrams - Adjusted Bandwidth (Small Measurements) vs Latency", fontsize=13)
ax2.set_xlabel("Bandwidth (Mbps)")
ax2.set_ylabel("Latency (seconds)")

handles, labels = ax2.get_legend_handles_labels()
ax2.legend(handles[::-1], labels[::-1])

all_data = np.loadtxt(open("datagrams/results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
plt.autoscale(True)
ax3 = fig.add_subplot(1,1,1)
ax3.plot(all_data[:,0], all_data[:,5])
ax3.set_title("Datagrams - Adjusted Bandwidth vs Loss", fontsize=13)
ax3.set_xlabel("Bandwidth (Mbps)")
ax3.set_ylabel("Loss (percentage)")

all_data = np.loadtxt(open("datagrams/small_results.csv", "rb"), delimiter=",", skiprows=1)
fig = plt.figure()
plt.autoscale(True)
ax3 = fig.add_subplot(1,1,1)
ax3.plot(all_data[:,0], all_data[:,5])
ax3.set_title("Datagrams - Adjusted Bandwidth (Small Measurements) vs Loss", fontsize=13)
ax3.set_xlabel("Bandwidth (Mbps)")
ax3.set_ylabel("Loss (percentage)")
```
